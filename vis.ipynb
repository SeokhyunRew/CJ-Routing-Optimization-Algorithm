```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rc
rc("font", family='AppleGothic')
```

```python
final_model = pd.read_csv("VEH_RESULT.csv")
TA = pd.read_csv("VEH_RESULT_fitcbm.csv")
NO = pd.read_csv("VEH_RESULT_comeback_near_o_rate0.4.csv")
base_model = pd.read_csv("VEH_RESULT_basecode.csv")
NO_0 = pd.read_csv("VEH_RESULT_comeback_near_o_not_rate.csv")
```

```python
주문처리건수_fm_std = final_model["Count"].std()
주문처리건수_ta_std = TA["Count"].std()
주문처리건수_no_std = NO["Count"].std()
주문처리건수_bm_std = base_model["Count"].std()
주문처리건수_no0_std = NO_0["Count"].std()

주문처리건수_fm_me = final_model["Count"].mean()
주문처리건수_ta_me = TA["Count"].mean()
주문처리건수_no_me = NO["Count"].mean()
주문처리건수_bm_me = base_model["Count"].mean()
주문처리건수_no0_me = NO_0["Count"].std()

print("Number of order per Vehicle")
print(f"final_model: {주문처리건수_fm_me} ,{주문처리건수_fm_std}")
print(f"truck assignment: {주문처리건수_ta_me}, {주문처리건수_ta_std}")
print(f"nearest origin: {주문처리건수_no_me}, {주문처리건수_no_std}")
print(f"base model: {주문처리건수_bm_me}, {주문처리건수_bm_std}")
print(f"nearest origin 0: {주문처리건수_no0_me}, {주문처리건수_no0_std}")
```

```python
# '컬럼1' 데이터로 히스토그램 그리기
# 격자 여백 설정
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

plt.subplot(221)
plt.hist(final_model['Count'], edgecolor='k', color="cyan")  # bins는 구간(bin)의 개수를 나타냅니다
plt.title("Final model")

plt.subplot(222)
plt.hist(TA['Count'], edgecolor='k')  # bins는 구간(bin)의 개수를 나타냅니다
plt.title("Vehicle Assignment")

plt.subplot(223)
plt.hist(NO['Count'], edgecolor='k')  # bins는 구간(bin)의 개수를 나타냅니다
plt.title("Nearest Origin")

plt.subplot(224)
plt.hist(base_model['Count'], edgecolor='k')  # bins는 구간(bin)의 개수를 나타냅니다
plt.title("Base model")
plt.show()
```

```python
name = ["Final model", "Vehicle Assignment", "Nearest Origin", "Base model"]
사용차량수 = []
사용차량수.append(len(final_model))
사용차량수.append(len(TA))
사용차량수.append(len(NO))
사용차량수.append(len(base_model))

colors = ["cyan", "steelblue", "steelblue", "steelblue"]

plt.bar(name, 사용차량수, color=colors, edgecolor='k')

for i, v in enumerate(사용차량수):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
